```python
import cv2
from google.colab import drive
drive.mount('/content/drive')
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
```

# Change image to GreyScale

```python
img = cv2.imread("/content/drive/My Drive/image2.jpg")
#cvtColor: Converts an image from one color space to another.
#firstParameter:input source,secondParamter:coding 

grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img)
a = np.array(grey).flatten()
ax = np.array([x for x in range(0, len(a))])
plt.scatter(ax,a,c='black')
```

# Median Filter

```python
#The medianBlur function from the Open-CV library can be used to implement a median filter.
blur_median = cv2.medianBlur(grey, 7)
cv2_imshow(blur_median)
a = np.array(blur_median).flatten()
ax = np.array([x for x in range(0, len(a))])
plt.scatter(ax,a,c='black')
```

# Edge Detection Using Laplacian Filter

```python
##Spatial filter,Isotropic use second derivative to find area of rapid change (Zero crossing edge detection)
after_laplacian = cv2.Laplacian(blur_median,cv2.CV_64F, ksize=5)
_,edges = cv2.threshold(after_laplacian,125,255,cv2.THRESH_BINARY_INV)
a = np.array(edges).flatten()
ax = np.array([x for x in range(0, len(a))])
plt.scatter(ax,a,c='black')
cv2_imshow(edges)

```

# Generating a color painting and a cartoon

```python
"""
d  Diameter of each pixel neighborhood that is used during filtering. If it is non-positive, it is computed from sigmaSpace .
sigmaColor  Filter sigma in the color space. A larger value of the parameter means that farther colors within the pixel
neighborhood (see sigmaSpace ) will be mixed together, resulting in larger areas of semi-equal color.
sigmaSpace  Filter sigma in the coordinate space. A larger value of the parameter means that
farther pixels will influence each other as long as their colors are close enough (see sigmaColor ).
When d>0 , it specifies the neighborhood size regardless of sigmaSpace . Otherwise, d is proportional to sigmaSpace 
"""
x=img
cv2_imshow(x)
```
